### Initialize the raw data

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from sklearn.metrics import r2_score
from matplotlib import ticker
import pandas as pd


U = [4.00*a/10.00 for a in range(0,33)]; U[0] = 0.02; U[31]=12.50; U[32]=13.00
U_err = [0.01/2 for a in range(0,33)]
I=[5,64,122,175,220,257,282,296,304,295,293,292,292,285,284,285,285,288,296,305,314,314,313,313,312,312,312,313,313,313,314,314,314]
I_err=[1/2 for a in range(0,33)]
U_cut=[4*a/10 for a in range(21,33)]; U_cut[10]=12.50; U_cut[11]=13.00
U_cut_err=[0.01/2 for a in range(21,33)]
P=[12.3,17.8,23.0,30.0,39.0,47.0,57.0,68.0,77.8,88.0,99.0,113.0]
P_err=[0.1/2 for a in range(21,33)]
f=[9.1350,9.1105,9.0900,9.0750,9.0615,9.0445,9.0335,9.0155,8.9975,8.9865,8.9765,8.9570]
f_err=[0.0005/2 for a in range(21,33)]


print(len(U))
print(len(I))
```

### Ready to plot it!

```python
#plt.scatter(x, y, label='data')
fig=plt.figure(figsize=(12,9))

ax1 = fig.add_subplot(111)
ax1.errorbar(U, I, xerr=U_err, yerr=I_err ,fmt='o', label=r'Working Current $\rm I$', color='blue', markersize=5, ecolor='blue', capsize=2)
ax1.legend(loc='upper left', frameon=False, fontsize=16)
ax1.set_ylabel(r'$\rm I\ /\ mA$', fontsize=16, color="blue") 
ax1.set_xlabel(r'$\rmU\ /\ V$', fontsize=16, color="black") 
ax1.grid()
ax1.set_ylim(0,500)
#ax1.yaxis.label.set_color('red') 
ax1.tick_params(axis='y', direction='out', colors='blue',labelsize=14) 
ax1.plot(U,I,'b--')

ax2 = ax1.twinx()
ax2.errorbar(U_cut, P, xerr=U_cut_err, yerr=P_err,fmt='o', label=r'Relative Power $\rm P$', color='green', markersize=5, ecolor='green', capsize=2)
ax2.legend(loc='lower right', frameon=False, fontsize=16)
ax2.set_ylim(0,150)
#ax2.set_xlim([0, np.e])
#ax2.set_xlabel('Same X for both exp(-x) and ln(x)')
ax2.set_ylabel(r"$\rm P$", fontsize=16, color="darkgreen")
#ax2.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=0))
ax2.tick_params(axis='y', direction='out',  colors='darkgreen',labelsize=14)
ax2.plot(U_cut,P,'g--')

ax3 = ax1.twinx()
ax3.errorbar(U_cut, f, xerr=U_cut_err, yerr=f_err,fmt='o', label=r'Microwave Frequency $\rm f$', color='red', markersize=5, ecolor='red', capsize=2)
ax3.legend(loc='upper right', frameon=False, fontsize=16)
ax3.set_ylim(8.8,9.2)
ax3.set_ylabel(r'$\rm f\ /\ GHz$', labelpad=-70, fontsize=16, color="red")
#ax3.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=0))
ax3.tick_params(axis='y',direction='in', pad=-40, colors='red',labelsize=14)
ax3.plot(U_cut,f,'r--')

x_major_locator=MultipleLocator(1)
y1_major_locator=MultipleLocator(50)
y2_major_locator=MultipleLocator(15)
y3_major_locator=MultipleLocator(0.04)

ax1.xaxis.set_major_locator(x_major_locator)
ax1.yaxis.set_major_locator(y1_major_locator)
ax2.yaxis.set_major_locator(y2_major_locator)
ax3.yaxis.set_major_locator(y3_major_locator)

#plt.title(r"$\rm P_{IR}-I$ Distribution & Conversion Efficiency from $\rm P_{LD}$ to $\rm P_{IR}$", fontsize=18)
#plt.xlabel(r'$\rmI\ /\ A$', fontsize=16, color="black") 
plt.grid(True)
#plt.legend(loc='best', frameon=False)
plt.xlim(0,15)
#plt.ylim(-5,110)
#plt.axis([30, 400, 100, 400])
#plt.plot(myline, mymodel(myline), color = 'green', linestyle='dashed', label='Linear Fit')
#plt.text(0.5, 1400,r"$y=%.5fx%.5f$"%(a1,a0), fontsize=16,color="green")
ax=plt.gca()
ax.spines['left'].set_color('blue')
ax.spines['right'].set_color('darkgreen')

plt.savefig('phyex1_fig.pdf')
plt.show()
```
