### Initialize the raw data

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from sklearn.metrics import r2_score
from matplotlib import ticker
import pandas as pd

l = [a*1.0 for a in range(88, 113)]
l[0]=87.8
l_err = [0.1/2 for a in range(88, 113)]

I=[0.0,1.6,5.4,12.1,20.8,30.6,41.0,52.0,60.4,68.2,74.3,78.0,78.9,76.9,74.0,68.0,60.5,51.8,42.0,32.0,22.9,14.0,7.4,2.2,0.0]
I_err = [0.1/2 for a in range(88, 113)]
E = [np.sin((l[a]-87.8)/48.85*2*np.pi) for a in range(113-88)]
E_err = [np.cos((l[a]-87.8)/48.85*2*np.pi)*2*np.pi*l_err[a]/48.85 for a in range(113-88)]

logI=[0 for a in range(113-88)]
logI_err=[0 for a in range(113-88)]
logE=[0 for a in range(113-88)]
logE_err=[0 for a in range(113-88)]
for a in range(1,113-88-1):
    logI[a]=np.log(I[a])
    logI_err[a]=I_err[a]/I[a]
    logE[a]=np.log(E[a])
    logE_err[a]=E_err[a]/E[a]

print(logI)
print(logE)
```

```python
l = [a*1.0 for a in range(89, 112)]
l_err = [0.1/2 for a in range(89, 112)]

I=[1.6,5.4,12.1,20.8,30.6,41.0,52.0,60.4,68.2,74.3,78.0,78.9,76.9,74.0,68.0,60.5,51.8,42.0,32.0,22.9,14.0,7.4,2.2]
I_err = [0.1/2 for a in range(89, 112)]
E = [np.sin((l[a]-87.8)/48.85*2*np.pi) for a in range(112-89)]
E_err = [np.cos((l[a]-87.8)/48.85*2*np.pi)*2*np.pi*l_err[a]/48.85 for a in range(112-89)]

logI=[0 for a in range(112-89)]
logI_err=[0 for a in range(112-89)]
logE=[0 for a in range(112-89)]
logE_err=[0 for a in range(112-89)]
for a in range(112-89):
    logI[a]=np.log(I[a])
    logI_err[a]=I_err[a]/I[a]
    logE[a]=np.log(E[a])
    logE_err[a]=E_err[a]/E[a]
```

```python
from sklearn.metrics import r2_score

p, cov=np.polyfit(logE, logI, 1, cov=True)
mymodel = np.poly1d(p)
myline = np.linspace(-2, 0.2, 100)
a1 = mymodel.coefficients[0]
a0 = mymodel.coefficients[1]
r2=r2_score(logI, mymodel(logE))
print("R2=%.4f"%r2)
# print("the values are %d %7.5f %s" % (x, y, z))
print(mymodel.coefficients)

print(np.sqrt(np.diag(cov)))
sigma1,sigma0=np.sqrt(np.diag(cov))
print(logE)
print(logI)
```

```python
##plt.scatter(x, y, label='data')
fig=plt.figure(figsize=(12,9))

ax1 = fig.add_subplot(111)
ax1.errorbar(logE, logI, xerr=logI_err, yerr=logE_err ,fmt='o', label=r'$\rm Data$', color='blue', markersize=5, ecolor='blue', capsize=2)
ax1.legend(loc=1, frameon=False, fontsize=16)
ax1.set_ylabel(r'$\rm logI(x)$', fontsize=16, color="black") 
ax1.set_xlabel(r'$\rm log\left|E\right|$', fontsize=16, color="black") 
ax1.grid()
ax1.set_ylim(-1,5)
##ax1.yaxis.label.set_color('red') 
#ax1.tick_params(axis='y', colors='blue') 
#ax1.plot(logE,logI,'b--')

x_major_locator=MultipleLocator(0.2)
y1_major_locator=MultipleLocator(1)

ax1.xaxis.set_major_locator(x_major_locator)
ax1.yaxis.set_major_locator(y1_major_locator)

plt.plot(myline, mymodel(myline), color = 'green', linestyle='dashed', label='Linear Fit')

#plt.title(r"$\rm Curve\ of\ Standing\ wave\ I(x)$", fontsize=18)
##plt.xlabel(r'$\rmI\ /\ A$', fontsize=16, color="black") 
plt.grid(True)
##plt.legend(loc='best', frameon=False)
plt.xlim(-2,0.2)
##plt.ylim(-5,110)
##plt.axis([30, 400, 100, 400])
##plt.plot(myline, mymodel(myline), color = 'green', linestyle='dashed', label='Linear Fit')
##plt.text(0.5, 1400,r"$y=%.5fx%.5f$"%(a1,a0), fontsize=16,color="green")
ax=plt.gca()
#ax.spines['left'].set_color('blue')
#ax.spines['right'].set_color('darkgreen')

plt.text(-1, 1.9,r"$\rm logI=%.5f*log\left|E\right|+%.5f$"%(a1,a0), fontsize=16,color="green")
plt.text(-1, 1.5, r"$a_1=%f\pm%f$"%(a1,sigma1), fontsize=16,color="green")
plt.text(-1, 1.1, r"$a_0=%f\pm%f$"%(a0,sigma0), fontsize=16,color="green")
plt.text(-1, 0.7, r"$R^2=%f$"%r2, fontsize=16, color="green")
plt.legend(loc=0, frameon=False, fontsize=16)

plt.savefig('phyex4_fig.pdf')
plt.show()
```
