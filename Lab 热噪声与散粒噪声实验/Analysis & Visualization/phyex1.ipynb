```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from sklearn.metrics import r2_score
from matplotlib import ticker
import pandas as pd
import plotly.express as px
from scipy.interpolate import CubicSpline

R=[100, 200, 500, 1e3, 10e3, 20e3, 50e3, 100e3]
sqrt_y=1e-9*np.array([1.4920, 2.0911,1.9594,2.9101,10.888,15.383,26.287,36.136])
y=sqrt_y**2
```

```python
p, cov=np.polyfit(R, y, 1, cov=True)
mymodel = np.poly1d(p)
linearFit = np.poly1d(p)
a1 = mymodel.coefficients[0]
a0 = mymodel.coefficients[1]
sigma1,sigma0=np.sqrt(np.diag(cov))
a1
```
