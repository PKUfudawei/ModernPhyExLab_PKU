```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from sklearn.metrics import r2_score
from matplotlib import ticker
import pandas as pd
from scipy.optimize import curve_fit


with open('/Users/DaweiFu/fdw/al_5th_0.05_234.0_数据文件.TXT','r') as f1: N1 = f1.read().splitlines()
with open('/Users/DaweiFu/fdw/al_5th_0.10_234.0_数据文件.TXT','r') as f2: N2 = f2.read().splitlines()
with open('/Users/DaweiFu/fdw/al_5th_0.15_234.0_数据文件.TXT','r') as f3: N3 = f3.read().splitlines()
with open('/Users/DaweiFu/fdw/al_5th_0.20_234.0_数据文件.TXT','r') as f4: N4 = f4.read().splitlines()

N1 = [ int(x)/600 for x in N1 ]
N2 = [ int(x)/600 for x in N2 ]
N3 = [ int(x)/600 for x in N3 ]
N4 = [ int(x)/600 for x in N4 ]

lnN1 = np.log(N1)
lnN2 = np.log(N2)
lnN3 = np.log(N3)
lnN4 = np.log(N4)

x=[a for a in range(1, 513)]
```

```python
lnN1_bkg=[lnN1[i] for i in range(15,100)]
x_bkg=[i for i in range(15,100)]

p, cov=np.polyfit(x_bkg, lnN1_bkg, 1, cov=True)
mymodel1 = np.poly1d(p)
myline = np.linspace(15, 100)
a11 = mymodel1.coefficients[0]
a10 = mymodel1.coefficients[1]

r2=r2_score(lnN1_bkg, mymodel1(x_bkg))
print("R2=%.6f"%r2)
# print("the values are %d %7.5f %s" % (x, y, z))
print(mymodel1.coefficients)

print(np.sqrt(np.diag(cov)))
sigma11,sigma10=np.sqrt(np.diag(cov))


plt.xlim(0, 200)
plt.ylim(-2, 1)
plt.scatter(x, lnN1, label='data', marker='o', color='black', s=1)
plt.plot(myline, mymodel1(myline), color = 'blue', linestyle='dashed', label='bkg fit')
plt.legend(loc=1, frameon=False, fontsize=16)
```

```python
N1_sig=[ N1[i] - np.exp(a11*i + a10) for i in range(0, 512) ]
N2_sig=[ N2[i] - np.exp(a21*i + a20) for i in range(0, 512) ]
N3_sig=[ N3[i] - np.exp(a31*i + a30) for i in range(0, 512) ]
N4_sig=[ N4[i] - np.exp(a41*i + a40) for i in range(0, 512) ]

for i in range(0, 100):
    N1_sig[i]=0
    N2_sig[i]=0
    N3_sig[i]=0
    N4_sig[i]=0

for i in range(0, 512):
    if np.exp(a11*(i+1) + a10)>N1[i]: N1_sig[i]=0
    if np.exp(a21*(i+1) + a20)>N2[i]: N2_sig[i]=0
    if np.exp(a31*(i+1) + a30)>N3[i]: N3_sig[i]=0
    if np.exp(a41*(i+1) + a40)>N4[i]: N4_sig[i]=0
    
print(np.sum(N1_sig))
print(np.sum(N2_sig))
print(np.sum(N3_sig))
print(np.sum(N4_sig))
```
