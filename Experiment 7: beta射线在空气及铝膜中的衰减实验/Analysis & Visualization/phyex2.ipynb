```python
N0_sig=[ N0[i] - np.exp(a01*i + a00) for i in range(0, 512) ]
N1_sig=[ N1[i] - np.exp(a11*i + a10) for i in range(0, 512) ]
N2_sig=[ N2[i] - np.exp(a21*i + a20) for i in range(0, 512) ]
N3_sig=[ N3[i] - np.exp(a31*i + a30) for i in range(0, 512) ]
N4_sig=[ N4[i] - np.exp(a41*i + a40) for i in range(0, 512) ]

for i in range(0, 130):
    N0_sig[i]=0
    N1_sig[i]=0
    N2_sig[i]=0
    N3_sig[i]=0
    N4_sig[i]=0

for i in range(0, 512):
    if np.exp(a01*(i+1) + a00)>N1[i]: N0_sig[i]=0
    if np.exp(a11*(i+1) + a10)>N1[i]: N1_sig[i]=0
    if np.exp(a21*(i+1) + a20)>N2[i]: N2_sig[i]=0
    if np.exp(a31*(i+1) + a30)>N3[i]: N3_sig[i]=0
    if np.exp(a41*(i+1) + a40)>N4[i]: N4_sig[i]=0
```
